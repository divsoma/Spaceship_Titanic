```python
train=train.drop(['PassengerId','Name','Cabin'],axis=1)
```

```python
train.isnull().sum()
```

```python
train=train.fillna(train.mean())
```

```python
train.isnull().sum()
```

```python
le=LabelEncoder()
train["HomePlanet"]=le.fit_transform(train['HomePlanet'])
train["CryoSleep"]=le.fit_transform(train['CryoSleep'])
train["Destination"]=le.fit_transform(train['Destination'])
train["VIP"]=le.fit_transform(train["VIP"])
train["Transported"]=le.fit_transform(train["Transported"])
```

```python
scale = StandardScaler()
train[['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']]=scale.fit_transform(train[['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']])
train.head()
```

```python
x=train.iloc[:,:-1]
y=train.iloc[:,-1]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
xgb=XGBClassifier()
xgb.fit(xtrain,ytrain)
ypred=xgb.predict(xtest)
```

```python
test=test.drop(['Name','Cabin'],axis=1)
test.head()
```

```python
le=LabelEncoder()
test["HomePlanet"]=le.fit_transform(test['HomePlanet'])
test["CryoSleep"]=le.fit_transform(test['CryoSleep'])
test["Destination"]=le.fit_transform(test['Destination'])
test["VIP"]=le.fit_transform(test["VIP"])
```

```python
scale = StandardScaler()
test[['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']]=scale.fit_transform(test[['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']])
```

```python
ypred=xgb.predict(test.drop(['PassengerId'], axis=1))
ypred=ypred.astype(bool)
ypred.shape
```

```python
x=pd.DataFrame(test['PassengerId'],columns=['PassengerId'])

y=pd.DataFrame(ypred)
submission=pd.concat([x,y],axis=1)
submission.shape
```

```python
submission.to_csv('sample_submission.csv',index=False)
```
